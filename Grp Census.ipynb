```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
tech = pd.read_csv("Copy of tech_characteristics_of_businesses.csv")
```

```python
pd.unique(tech['FACTORS_P_LABEL'])
```

```python
tech['FACTORS_P_LABEL'].value_counts()
```

```python
plt.rcParams['figure.figsize'] = [15, 5]
```

```python
ax = tech.sort_values(by = 'FACTORS_P_LABEL', ascending=False)['FACTORS_P_LABEL'].value_counts()[1:10].plot(x='FACTORS_P_LABEL', kind="barh")
plt.title('Instances of issues pertaining to certain technologies')
plt.xlabel('Adoption Factors')
plt.ylabel('Instances')
plt.xticks(rotation = 90)
plt.show()
```
